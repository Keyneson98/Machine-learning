```python
import matplotlib.pyplot as plt
import pandas 
import time
import psutil
```

```python
%matplotlib notebook
plt.rcParams['animation.html']= 'jshtml'
```

```python
fig = plt.figure()
ax= fig.add_subplot(111)
fig.show
```

```python
i=0
x,y= [],[]
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    
    ax.plot(x,y)
    fig.canvas.draw()
    
    ax.set_xlim(left=max(0, i-50), right= i+50)
    
    time.sleep(0.1)
    i += 1
```
